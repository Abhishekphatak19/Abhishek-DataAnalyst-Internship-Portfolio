### Data Loading

The dataset is loaded into a Pandas DataFrame for analysis.  
Initial inspection of the dataset helps understand its structure, including the number of rows and columns.

```python
# Load dataset
df = pd.read_csv("/content/online_retail.csv")
print("Shape:", df.shape)
df.head()
```

```python
import pandas as pd
from scipy.stats import ttest_ind

# Ensure datetime
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format="%d-%m-%Y %H:%M")

# Create total price (if not already created)
df['total price'] = df['Quantity'] * df['UnitPrice']
```

```python
uk = df[df['Country'] == 'United Kingdom']['total price']
non_uk = df[df['Country'] != 'United Kingdom']['total price']
```

```python
t_stat, p_value = ttest_ind(uk, non_uk, equal_var=False)

print("T-Statistic:", t_stat)
print("P-Value:", p_value)
```

```python
if p_value < 0.05:
    print("Significant difference → Reject Null Hypothesis")
else:
    print("No significant difference → Fail to Reject Null")
```

```python
df['order weekday'] = df['InvoiceDate'].dt.day_name()
df['is weekend'] = df['order weekday'].isin(['Saturday', 'Sunday'])
```

```python
weekend = df[df['is weekend'] == True]['total price']
weekday = df[df['is weekend'] == False]['total price']
```

```python
t_stat, p_value = ttest_ind(weekend, weekday, equal_var=False)

print("P-Value:", p_value)
```

```python
correlation = df['Quantity'].corr(df['total price'])

print("Correlation:", correlation)
```

```python
from scipy.stats import chi2_contingency

# Create category
df['high sales'] = df['total price'] > df['total price'].median()

# Contingency table
table = pd.crosstab(df['Country'], df['high sales'])

chi2, p, dof, expected = chi2_contingency(table)

print("P-Value:", p)
```
